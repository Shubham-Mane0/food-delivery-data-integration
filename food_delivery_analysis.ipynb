# Food Delivery Data Integration

This notebook loads data from CSV, JSON, and SQL files, merges them, and creates a final dataset.

```python
import pandas as pd
import json
import sqlite3
```

## Load Orders Data (CSV)

```python
orders_df = pd.read_csv('orders.csv')
orders_df.head()
```

## Load Users Data (JSON)

```python
with open('users.json', 'r') as f:
    users_data = json.load(f)

users_df = pd.DataFrame(users_data)
users_df.head()
```

## Load Restaurants Data (SQL)

```python
conn = sqlite3.connect(':memory:')
with open('restaurants.sql', 'r') as f:
    sql_script = f.read()
conn.executescript(sql_script)
restaurants_df = pd.read_sql_query('SELECT * FROM restaurants', conn)
conn.close()

restaurants_df.head()
```

## Merge All Data

```python
merged_df = pd.merge(orders_df, users_df, on='user_id', how='left')
final_df = pd.merge(merged_df, restaurants_df, on='restaurant_id', how='left', suffixes=('', '_info'))
final_df.head()
```

## Save Final Dataset

```python
final_df.to_csv('final_food_delivery_dataset.csv', index=False)
print("Final dataset 'final_food_delivery_dataset.csv' has been created successfully.")
```
